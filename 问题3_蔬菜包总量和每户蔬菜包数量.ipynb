```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
sns.set_context("paper")
plt.rcParams["font.sans-serif"]='KaiTi'   #解决中文乱码问题
plt.rcParams['axes.unicode_minus']=False   #解决负号无法显示的问题
%matplotlib inline
```

### 可视化每个区每天蔬菜包的发放数量

```python
df_1 = pd.read_excel('./蔬菜包汇总.xlsx')
df_1
```

```python
df_2 = df_1.iloc[:-1, :]
df_2
```

```python
plt.figure(figsize=(12, 6))
plt.title('不同区蔬菜包分发情况',fontsize=16)
plt.xlabel("日期", fontsize=16)
plt.ylabel("蔬菜包件数", fontsize=16)
markers = ['-*', '-o', '-^', '-+', '-x', '-+', '-o', '-^', '-p', '-d']
location = df_2.columns
for i in range(1, 12):
    marker = markers[i%len(markers)]
    plt.plot(df_2['时间'], df_2[location[i]], marker,label=location[i])
plt.legend()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('./images/问题3_不同区蔬菜包分发情况.png')
plt.show()
```

### 可视化每个区的总蔬菜包量和人均蔬菜包量

```python
df_3 = df_1.iloc[-1, 1:]
# 每个区的户数
df_4 = pd.read_excel('./小区统计/2.xlsx')
df_4['小区户数（户）']
```

```python
df_3
```

```python
plt.figure(figsize=(15,6))
plt.title('长春市各区蔬菜包总量和每户蔬菜包数量', fontsize=16)
xticks = np.arange(9) 
bar_width = 0.3
plt.bar(xticks, df_3[df_4['所属区域']], width=bar_width, label='总蔬菜包数量')
for a, b in zip(xticks, df_3[df_4['所属区域']]):
    plt.text(a, b + 2, b, ha='center', va='bottom')
plt.bar(xticks, (df_3[df_4['所属区域']].values/df_4['小区户数（户）'].values), width=bar_width, label='户均蔬菜包数量')
for a, b in zip(xticks+bar_width, (df_3[df_4['所属区域']].values/df_4['小区户数（户）'].values)):
    plt.text(a, b + 2, "%.2f" % b, ha='center', va='bottom')
plt.xticks(xticks,df_4['所属区域'],size=16)
plt.xlabel('区域名称', fontsize=16)
plt.ylabel('蔬菜包数量/包', fontsize=16)
plt.legend()
plt.savefig('./images/问题3_户均蔬菜包数量.png')
plt.show()
```

从人均蔬菜包数量来看，长春新区、南关区和朝阳区供应量不充足。
